```python

from google.colab import files
uploaded = files.upload()

import pandas as pd
df = pd.read_csv("labours_dataset_10000.csv")
print("Dataset loaded, first 5 rows:")
print(df.head())

# 3️⃣ Combine text columns for better matching
df['combined_text'] = df['skills_text'] + " " + df['text']

# 4️⃣ Vectorize text (TF-IDF)
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000)
X = vectorizer.fit_transform(df['combined_text'])
print("Text vectorization done!")

# 5️⃣ Train Nearest Neighbors model (for recommendations)
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=5, metric='cosine')
model.fit(X)
print("NearestNeighbors model trained!")

# 6️⃣ Test model with example requirement
requirement = "Need tile worker with 5+ years in Kurunegala"
req_vec = vectorizer.transform([requirement])
distances, indices = model.kneighbors(req_vec)

print("\nTop 5 matching labours for requirement:")
for i in indices[0]:
    print(df.iloc[i][['title', 'skills_text', 'location', 'experience', 'feePerHour']])

# 7️⃣ Save model & vectorizer
import joblib
joblib.dump(model, "labour_model.pkl")
joblib.dump(vectorizer, "tfidf_vectorizer.pkl")
print("\nModel and vectorizer saved!")

# 8️⃣ Download files
files.download("labour_model.pkl")
files.download("tfidf_vectorizer.pkl")
```
